```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error,r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input, Conv1D, GlobalMaxPooling1D, Concatenate
from tensorflow.keras.optimizers import Adam
```

```python
# Assuming 'close' is the target variable
target = df['close'].values
```

```python
# Select relevant features for quantitative and qualitative data
quant_features = df[['high', 'low', 'volume', 'market_cap']].values
qual_features = df[['reddit_posts', 'reddit_posts_score', 'youtube', 'social_volume','social_score','average_sentiment','news','price_score','social_impact_score','correlation_rank','galaxy_score','volatility','market_cap_rank','market_cap_global']].values
```

```python
# Normalize the quantitative features
scaler = MinMaxScaler()
quant_features_scaled = scaler.fit_transform(quant_features)
```

```python
# Normalize the qualitative features
qual_scaler = MinMaxScaler()
qual_features_scaled = qual_scaler.fit_transform(qual_features)
```

```python
# Split the data into training and testing sets
split = int(0.8 * len(df))
quant_train, quant_test = quant_features_scaled[:split], quant_features_scaled[split:]
qual_train, qual_test = qual_features_scaled[:split], qual_features_scaled[split:]
target_train, target_test = target[:split], target[split:]
```

```python
# Define the LSTM model for quantitative data
input_quant = Input(shape=(quant_train.shape[1], 1))
x = LSTM(64, return_sequences=True)(input_quant)
x = LSTM(32)(x)
x = Dropout(0.2)(x)
output_quant = Dense(1)(x)
```

```python
# Define the CNN model for qualitative data
input_qual = Input(shape=(qual_train.shape[1], 1))
y = Conv1D(32, kernel_size=3, activation='relu')(input_qual)
y = GlobalMaxPooling1D()(y)
y = Dense(32, activation='relu')(y)
output_qual = Dense(1)(y)
```

```python
# Concatenate outputs from LSTM and CNN
combined = Concatenate()([output_quant, output_qual])
z = Dense(64, activation='relu')(combined)
z = Dropout(0.2)(z)
final_output = Dense(1)(z)

# Define the final model
model = Model(inputs=[input_quant, input_qual], outputs=final_output)
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
# Reshape data for LSTM and CNN models
quant_train_reshaped = quant_train.reshape((quant_train.shape[0], quant_train.shape[1], 1))
quant_test_reshaped = quant_test.reshape((quant_test.shape[0], quant_test.shape[1], 1))
qual_train_reshaped = qual_train.reshape((qual_train.shape[0], qual_train.shape[1], 1))
qual_test_reshaped = qual_test.reshape((qual_test.shape[0], qual_test.shape[1], 1))
```

```python
# Train the model
model.fit([quant_train_reshaped, qual_train_reshaped], target_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Predict on test set
predictions = model.predict([quant_test_reshaped, qual_test_reshaped])

mse = mean_squared_error(target_test, predictions)
mae = mean_absolute_error(target_test, predictions)
# Calculate MAPE manually
mape = np.mean(np.abs((target_test - predictions) / np.where(target_test == 0, 1, target_test))) * 100
rmse = np.sqrt(mse)
r2 = r2_score(target_test, predictions)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Absolute Percentage Error: {mape}%")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

```python
from tensorflow.keras.layers import GRU

# Define the GRU model for quantitative data
input_quant_gru = Input(shape=(quant_train.shape[1], 1))
x_gru = GRU(64, return_sequences=True)(input_quant_gru)
x_gru = GRU(32)(x_gru)
x_gru = Dropout(0.2)(x_gru)
output_quant_gru = Dense(1)(x_gru)
```

```python
# Define the CNN model for qualitative data
input_qual_cnn = Input(shape=(qual_train.shape[1], 1))
y_cnn = Conv1D(32, kernel_size=4, activation='relu')(input_qual_cnn)
y_cnn = GlobalMaxPooling1D()(y_cnn)
y_cnn = Dense(32, activation='relu')(y_cnn)
output_qual_cnn = Dense(1)(y_cnn)
```

```python
# Concatenate outputs from GRU and CNN
combined_gru_cnn = Concatenate()([output_quant_gru, output_qual_cnn])
z_gru_cnn = Dense(64, activation='relu')(combined_gru_cnn)
z_gru_cnn = Dropout(0.2)(z_gru_cnn)
final_output_gru_cnn = Dense(1)(z_gru_cnn)

# Define the final model
model_gru_cnn = Model(inputs=[input_quant_gru, input_qual_cnn], outputs=final_output_gru_cnn)
model_gru_cnn.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
# Reshape data for GRU and CNN models
quant_train_reshaped = quant_train.reshape((quant_train.shape[0], quant_train.shape[1], 1))
quant_test_reshaped = quant_test.reshape((quant_test.shape[0], quant_test.shape[1], 1))
qual_train_reshaped = qual_train.reshape((qual_train.shape[0], qual_train.shape[1], 1))
qual_test_reshaped = qual_test.reshape((qual_test.shape[0], qual_test.shape[1], 1))
```

```python
# Train the model
model.fit([quant_train_reshaped, qual_train_reshaped], target_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Predict on test set
predictions = model.predict([quant_test_reshaped, qual_test_reshaped])

mse = mean_squared_error(target_test, predictions)
mae = mean_absolute_error(target_test, predictions)
# Calculate MAPE manually
mape = np.mean(np.abs((target_test - predictions) / np.where(target_test == 0, 1, target_test))) * 100
rmse = np.sqrt(mse)
r2 = r2_score(target_test, predictions)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Absolute Percentage Error: {mape}%")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

```python
from tensorflow.keras.layers import Bidirectional

# Define the Bi-LSTM model for quantitative data
input_quant_bilstm = Input(shape=(quant_train.shape[1], 1))
x_bilstm = Bidirectional(LSTM(64, return_sequences=True))(input_quant_bilstm)
x_bilstm = Bidirectional(LSTM(32))(x_bilstm)
x_bilstm = Dropout(0.2)(x_bilstm)
output_quant_bilstm = Dense(1)(x_bilstm)
```

```python
# Define the CNN model for qualitative data
input_qual_cnn2 = Input(shape=(qual_train.shape[1], 1))
y_cnn2 = Conv1D(64, kernel_size=5, activation='relu')(input_qual_cnn2)
y_cnn2 = GlobalMaxPooling1D()(y_cnn2)
y_cnn2 = Dense(32, activation='relu')(y_cnn2)
output_qual_cnn2 = Dense(1)(y_cnn2)
```

```python
# Concatenate outputs from Bi-LSTM and CNN
combined_bilstm_cnn = Concatenate()([output_quant_bilstm, output_qual_cnn2])
z_bilstm_cnn = Dense(64, activation='relu')(combined_bilstm_cnn)
z_bilstm_cnn = Dropout(0.2)(z_bilstm_cnn)
final_output_bilstm_cnn = Dense(1)(z_bilstm_cnn)

# Define the final model
model_bilstm_cnn = Model(inputs=[input_quant_bilstm, input_qual_cnn2], outputs=final_output_bilstm_cnn)
model_bilstm_cnn.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```
